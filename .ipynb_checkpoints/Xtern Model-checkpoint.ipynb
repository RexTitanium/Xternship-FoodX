```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
```

```python
X = df.drop(['Order'],axis = 1)

Y = df['Order']

X_train,X_test,Y_train,Y_test = train_test_split(X,Y, test_size=0.2, random_state = 5)
```

```python
model = RandomForestClassifier(n_estimators = 1500, random_state = 0)
```

```python
import pickle

def train_save(x_train, y_train, model, filename):
    model.fit(x_train, y_train)
    pickle.dump(model, open(filename,'wb'))
    print(f"Model trained and saved as {filename}")
```

```python
train_save(X_train, Y_train, model, 'model.pkl')
```

```python
loaded_model = pickle.load(open('model.pkl', 'rb'))
result = loaded_model.score(X_test, Y_test)
print(result)
```
